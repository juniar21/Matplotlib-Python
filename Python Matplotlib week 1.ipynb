```python
#Import Modules
    
import matplotlib.pyplot as plt  
import numpy as np
```

```python
#Membuat sample data

x = [1, 2, 3, 4]
y = [1, 2, 5, 4]

fig, ax = plt.subplots() #membuat sebuah figure dan sebuah axes
ax.plot(x, y) #melakukan plottin pada data axes
```

```python
#Alternatif lain visualisasi data menggunakan method plot() pada pyplot()

x = [1, 2, 3, 4]
y = [1, 2, 5, 4]

plt.plot(x,y)
```

```python
# Beberapa cara untuk membuat figure dan axes :

fig = plt.figure()  # figure tanpa axes
fig, ax = plt.subplots() #sebuah figure dengan sebuah axes
fig, axs = plt.subplots(2, 3) #sebuah figure dengan 2 x 3 grid axes
```

```python
# Dua cara dalam menggunakan matplotlib

1. OO Style (Object Oriented Style) 
2. pyplot Style
```

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots() #membuat sebuah figure dan sebuah axes

# plotting tiga variant data pada axes
ax.plot(x, x,  label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label') #menyertakan x-label pada axes
ax.set_ylabel('y label') #menyertakan y-label pada axes
ax.set_title("Simple Plot") #menyertakan title pada axes
ax.legend() #menyertakan legend
```
